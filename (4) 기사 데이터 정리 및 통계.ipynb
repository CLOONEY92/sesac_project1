```python
import pandas as pd
import csv
```

```python
#주가데이터 로드
stock = pd.read_csv('stock.csv', header=None, sep=',')
#칼럼설정
stock.columns=['날짜','주가']
#인덱스 설정
stock.index = stock['날짜']
```

```python
#삼성 데이터 로드
news = pd.read_csv('samsung.csv', header=None, sep=',')
#칼럼설정
news.columns=['날짜', '타이틀', '카테고리', '반응도', '본문']
#날짜 기준으로 기사개수 카운트
news_count = news.groupby(['날짜']).count()['본문']
#시리즈 => 데이터프레임으로 변경
news_count = pd.DataFrame(news_count)
```

```python
news_count
```

```python
#뉴스 데이터에 주가데이터 연결
news_count['주가'] = stock['주가']
```

```python
news_count
```

```python
#주가가 없는 날짜 제거 (주말)
news_count = news_count.dropna()
```

```python
news_count
```

```python
#백분위를 구하기 위한 데이터 합산
count_sum = news_count.sum()['본문']
stock_sum = news_count.sum()['주가']
```

```python
#백분위 구히기
news_count['본문%'] = news_count['본문'] / count_sum * 100
news_count['삼성전자주가%'] = news_count['주가'] / stock_sum * 100
```
