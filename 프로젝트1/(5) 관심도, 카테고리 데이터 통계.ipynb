```python
import pandas as pd
```

```python
#삼성 데이터 로드
news = pd.read_csv('samsung.csv', header=None, sep=',')
#칼럼설정
news.columns=['날짜', '타이틀', '카테고리', '반응도', '본문']
```

```python
#카테고리 기준으로 데이터 개수 합산
category = news.groupby('카테고리').count()['본문']
```

```python
category
```

```python
#반응도 기준으로 데이터 합산
reactivity = news.groupby('날짜').sum('반응도')
```
